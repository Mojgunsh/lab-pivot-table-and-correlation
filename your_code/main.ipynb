#### What's the mean salary by department?

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
data2=employee.pivot_table(index=['Department'],values=['Name','Salary'],aggfunc={'Name':'count','Salary':'sum'}).reset_index()
data2 = data2.rename(columns={'Name':'Number_employee'})
data2
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
data3=employee.pivot_table(index=['Department','Title'],values=['Years','Salary'],aggfunc={'Years':'sum','Salary':'mean'}).reset_index()
data3 = data3.rename(columns={'Salary':'AVG(salay)','Years':'Total year of experience'})
data3
```
